```python
def new_tiles(board):
    new_board = board.copy()
    
    open_indexes = []
    for i in range (4):
        for j in range(4):
             if board[i][j] == 0:
    
                 open_indexes.append(i)
                 open_indexes.append(j)
    open_indexes = numpy.reshape(open_indexes, (-1,2))
    num_open = len(open_indexes)

    rand1 = random.randint(0,num_open - 1)
    
    
    index1 = open_indexes[rand1]


    if (random.randint(0,9) == 0):
        value1 = 4
    else:
        value1 = 2
    

   
    new_board[index1[0]][index1[1]] = value1

    return(new_board)
```

```python
def compare_boards(board1, board2):
    for i in range(4):
        for j in range(4):
            if (board1[i][j] != board2[i][j]):
                return False
    return True
```

```python
def move_left(board,score):
    
    new_board = board.copy()
    
  
    for i in range(0,4):
        for j in range(1,4):
            
            
            if new_board[i][j-1] == new_board[i][j]:
                new_board[i][j-1] = new_board[i][j-1] *2
                score = score + new_board[i][j-1]
                new_board[i][j] = 0
                    
        
            
    
        
    for _ in range(4):
        initial_board = board.copy()
        for i in range(3,-1,-1):
            for j in range(3,0,-1):
                if new_board[i][j-1] == 0:
                    new_board[i][j-1] = new_board[i][j]
                    new_board[i][j] = 0
        if compare_boards(initial_board, new_board):
            return (new_board, score)
    return (new_board,score)
    

    
```

```python
def move_right(board,score):
    
    new_board = board.copy()
    
    for i in range(3,-1,-1):
        for j in range(2,-1,-1):
            if new_board[i][j+1] == new_board[i][j]:
                new_board[i][j+1] = new_board[i][j+1] * 2
                score = score + new_board[i][j+1] 
                new_board[i][j] = 0
                

    for _ in range(4):
        initial_board = board.copy()
        for i in range(0,4):
            for j in range(0,3):
                if new_board[i][j+1] == 0:
                    new_board[i][j+1] = new_board[i][j]
                    new_board[i][j] = 0
        if compare_boards(initial_board, new_board):
            return (new_board, score)
    return (new_board,score)
    
```

```python
def move_up(board,score):
    
    new_board = board.copy()
  
    for i in range(1,4):
        for j in range(0,4):
            if new_board[i-1][j] == new_board[i][j]:
                new_board[i-1][j] = new_board[i-1][j] * 2
                score = score + new_board[i-1][j] 
                new_board[i][j] = 0
        

    for _ in range(4):
        initial_board = board.copy()
        for i in range(3,0,-1):
            for j in range(3,-1,-1):
                if new_board[i-1][j] == 0:
                    new_board[i-1][j] = new_board[i][j]
                    new_board[i][j] = 0
        if compare_boards(initial_board, new_board):
            return (new_board, score)
    return (new_board,score)
```

```python
def move_down(board,score):
    
    new_board = board.copy()
  
    for i in range(2,-1,-1):
        for j in range(3,-1,-1):
            if new_board[i+1][j] == new_board[i][j]:
                new_board[i+1][j] = new_board[i+1][j] * 2
                score = score + new_board[i+1][j] 
                new_board[i][j] = 0
        
    for _ in range(4):
        initial_board = board.copy()
        for i in range(0,3):
            for j in range(0,4):
                if new_board[i+1][j] == 0:
                    new_board[i+1][j] = new_board[i][j]
                    new_board[i][j] = 0
        if compare_boards(initial_board, new_board):
            return (new_board, score)
    return (new_board, score)
```

```python
def game_over(board):
    
    game_state1 = move_left(board.copy(), 0)[0]
    game_state2 = move_right(board.copy(), 0)[0]
    game_state3 = move_up(board.copy(), 0)[0]
    game_state4 = move_down(board.copy(), 0)[0]
    
    if compare_boards (game_state1, game_state2):
        if compare_boards(game_state2, game_state3):
            if compare_boards(game_state3, game_state4):
                return True
    return False
```

```python
def translate_board(board):
    
    
            
    
    new_board = board.copy()
    
    largest_value = 0
    for i in range(4):
        for j in range(4):
            if new_board[i][j] > largest_value:
                largest_value = new_board[i][j]
                
    base2val = math.log(largest_value, 2)
            

    
    for i in range(4):
        for j in range(4):
            
            if new_board[i][j] > 0:
            
                num = math.log(new_board[i][j],2)
            else:
                num = 0
            new_board[i][j] = num / base2val
            
    return new_board

```

```python
def create_children(parent1, parent2, network1, network2, hiddenLayer1, hiddenLayer2):
    
    parent_weights1 = parent1.get_weights(network1)
    parent_weights2 = parent2.get_weights(network2)
    
    hidden_layer_weights1 = parent1.get_weights(hiddenLayer1)
    hidden_layer_weights2 = parent2.get_weights(hiddenLayer2)

    crossOverIndex = random.randint(0,64)
    tf.reset_default_graph()
    network = input_data(shape=[None, 4, 4], name='input')
    network = fully_connected(network, 4, activation='sigmoid')
    weights1 = numpy.zeros((16,4))
    for i in range(16):
        for j in range(4):
            
            if (i+1) + (16*j) < crossOverIndex: 
                value = hidden_layer_weights1[i][j] + float(random.randrange(-10,10)/50)
            else:
                value = hidden_layer_weights2[i][j] + float(random.randrange(-10,10)/50)
                
            rand = random.randint(0,19)
            if rand == 0:
                value = float(random.randrange(-100, 100))/100
            weights1[i][j] = value

    hidden_layer = network.W
    network = fully_connected(network, 4, activation='softmax')
    network = regression(network, optimizer='adam', learning_rate=1e-2, loss='mean_square', name='target')
    model = tflearn.DNN(network)

    crossOverIndex = random.randint(0,16)
    weights2 = numpy.zeros((4,4))
    for i in range(4):
        for j in range(4):
            if (i+1) + (16*j) < crossOverIndex: 
                value = parent_weights1[i][j] + float(random.randrange(-10,10)/50)
            else:
                value = parent_weights2[i][j] + float(random.randrange(-10,10)/50)
            rand = random.randint(0,19)
            if rand == 0:
                value = float(random.randrange(-100, 100))/100
            weights2[i][j] = value
    model.set_weights(network.W, weights2)
    model.set_weights(hidden_layer, weights1)
    

    return model, network.W, hidden_layer
    
```

```python
def choose_parents(fitness_percentages):
    num = float(random.randrange(0,10000)/100)
    num2 = float(random.randrange(0,10000)/100)
   
    parent_index1 = 0
    parent_index2 = 0
    done = False
    done2 = False
    for i in range(25):
        num = num - fitness_percentages[i]
        num2 = num2 - fitness_percentages[i]
        if num < 0 and done == False:
            
            parent_index1 = i
            done = True
        if num2 < 0 and done2 == False:
            
            parent_index2 = i
            done2 = True
        if done and done2:
            return (parent_index1, parent_index2)
            
    
    return (parent_index1, parent_index2)
    
```

```python
def new_population(best_models, best_networks, best_hiddenLayers, fitness_percentages):
    
    new_models = []
    new_networks = []
    new_hiddenLayers = []
    for i in range(75):
        #print(i)
        parent_indexes = choose_parents(fitness_percentages)
        parent1 = best_models[parent_indexes[0]]
        parent2 = best_models[parent_indexes[1]]
        networks1 = best_networks[parent_indexes[0]]
        networks2 = best_networks[parent_indexes[1]]
        hidden_layer1 = best_hiddenLayers[parent_indexes[0]]
        hidden_layer2 = best_hiddenLayers[parent_indexes[1]]
        
        child = create_children(parent1, parent2, networks1, networks2, hidden_layer1, hidden_layer2)
        model = child[0]
        network = child[1]
        hidden_layer = child[2]
        new_models.append(model)
        new_networks.append(network)
        new_hiddenLayers.append(hidden_layer)
    
    for i in range(25):
        new_models.append(best_models[i])
        new_networks.append(best_networks[i])
        new_hiddenLayers.append(best_hiddenLayers[i])
        
    return (new_models, new_networks, new_hiddenLayers)
        
```
